# Texture characterization using Haralick's GLCM

The example shows the utility of GLCM for texture characterisation. It is an easy and useful technique to extract relevant features from a texture, allowing us to implement systems for automated classification of texture images.



```python
import matplotlib.pyplot as plt
import numpy as np
import math
from skimage import data, io, filters, img_as_float
from skimage.feature import greycomatrix, greycoprops
from mpl_toolkits.mplot3d import Axes3D
from sklearn.metrics import mean_squared_error
```

# 1. Sample textures

Loading the images used as examples during the exercises.

```python
# Loading the textures
text1 = io.imread(fname="texture1.gif")
text2 = io.imread(fname="texture2.gif")
text3 = io.imread(fname="texture3.gif")

# Showing the textures
f, (ax0, ax1, ax2) = plt.subplots(ncols=3, figsize=(12, 12))
ax0.imshow(text1, cmap='gray', aspect='equal')
ax1.imshow(text2, cmap='gray', aspect='equal')
ax2.imshow(text3, cmap='gray', aspect='equal')
plt.tight_layout()
plt.show()
```

# 2. Function for GLCM computation

A function to compute the GLCM of a texture image is implemented. It involves two operations:

- The intensity values of the image are quantified into Ng levels.
- The quantified image is used as input for the cmputation of the GLCM atrix

# Tienes que escoger valores para los parmetros ng, d, y theta.

```python
def getGlcmMatrix(im, ng = , d = [], theta = []):
    # Quantification
    step = math.ceil(256.0/float(ng))
    imq = np.uint8(np.digitize(im, np.arange(0, 256, step))) - 1
    # Computing the GCM matrix for the quantified image
    glcm_matrix = greycomatrix(imq, d, theta, ng)
    return(glcm_matrix)
    
```

# 3. Getting GLCM matrices for our images

For each of the three textures analysed in the exercise, the GLCM is obtained by using ng levels of quantification, a distance of d for the comparision between pixels and an angle of theta to identify the neighbour pixel.

```python
glcm1=getGlcmMatrix(text1)
glcm2=getGlcmMatrix(text2)
glcm3=getGlcmMatrix(text3)
```

```python
print(glcm1.shape)
print(np.size(glcm1))
```
